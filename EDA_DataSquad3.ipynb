```python
import os
import pandas as pd
```

```python
#load dataset with electives
elective_df = pd.read_excel("tekems_section_demand_20230124_1930.xlsx")
print(elective_df.shape)
elective_df.head()
```

```python
#load dataset with student information
student_df = pd.read_excel("tekems_section_demand_by_student_20230126_1930_Com Chave.xlsx")
print(student_df.shape)
student_df.head()
```

```python
# merge the datasets on course column
data_all_df = elective_df.merge(student_df, on=['course', 'lective', 'section_term', 'section'] , how='left')
data_all_df.shape
```

```python
data_all_df.head()
```

```python
# filter for students from Management Master 
management_df = data_all_df[data_all_df["student_program_name"] == "Masters in Management"].reset_index(drop = True)
management_df.shape
```

```python
management_df.head()
```

```python
management_df.columns
```

```python
# delete double columns
management_df = management_df.drop(columns = ["last_section_update_y", "max_section_students_y", "section_vacancies_y"])

# rename columns
management_df = management_df.rename(columns = {"course_name_x" : "course_name_elective", "course_name_y" : "course_name_student", 
                               "last_section_update_x" : "last_section_update", "max_section_students_x" : 
                                "max_section_students", "section_vacancies_x" : "section_vacancies"})
```

```python
management_df.describe()
```

### Looking for missing values
As seen below missing values appear only in 4 columns.

Bidding group: 
- G1 = international students
- G2 = regular MSc students

Missing values appear when "list" is value "WAITLIST" or "CANCEL". Missing values will be replaced by 0.

For student allocated, bid, section_preference the values are missing because the "list" value is "CANCEL", so there was no allocation/bid/preference. Missing values will be replaced by value 2.

```python
missing_values = management_df.isna().sum().sort_values(ascending=False)
missing_values
```

```python
# replace missing values as described above
management_df = management_df.fillna({'bidding_group': 0, 'student_allocated': 2, 'bid': 2, 'section_preference' : 2})
```
